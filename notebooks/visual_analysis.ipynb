```python
import dill
import numpy as np
import matplotlib.pyplot as plt
```

```python
with open("data.dill", "rb") as f:
  data = dill.load(f)
y = data[1]
thr=0.9 #threshold to get indices with highest concentration values
marked_= np.where(y > thr)[0]
marked=marked_[:3]
print(marked)
```

The feature importance scores were calculated for only 100 windows and are visualised below

VISUALIZING FEATURE IMPORTANCE SCORES FOR RNN MODEL AND SYNTHETIC DATASET

```python
with open('rnn_shap_values.dill', 'rb') as f:
  shap_values=dill.load(f)
s=shap_values.reshape(shap_values.shape[0]*shap_values.shape[1], shap_values.shape[2])
marked_=range(0, s.shape[0], 250)
plt.figure(figsize=(50, 50))
fig, ax=plt.subplots(6)

for i in range(s.shape[1]):
  ax[i].plot(s[:, i])
  for mark in marked_:
      ax.flatten()[i].axvline(x=mark, color='r', linestyle='solid', linewidth=10, alpha=0.5)
for i in range(6):
  ax[i].set_xticks([])
  ax[i].set_xticklabels([])
  for a in ax:
    a.tick_params(axis='both', which='both', labelsize=7)
```

```python
plt.figure(figsize=(50, 50))
fig, ax=plt.subplots(7)

y_=y.reshape(y.shape[0], 1)


ax[0].plot(y_, linewidth=2)
# ax[0].vlines(marked-350, ymin=0, ymax=0.5,  color='r', linestyle='solid', linewidths=10, alpha=0.25)
for mark in marked:
  ax.flatten()[0].axvline(x=mark, color='r', linestyle='solid', linewidth=15, alpha=0.25)

for i in range(1, s.shape[1]+1):
  ax[i].plot(s[:, i-1], linewidth=2)
  for mark in marked:
    ax.flatten()[i].axvline(x=mark, color='r', linestyle='solid', linewidth=15, alpha=0.25)


# Remove x-axis ticks and labels
for i in range(7):
  ax[i].set_xticks([])
  ax[i].set_xticklabels([])


  for a in ax:
    a.tick_params(axis='both', which='both', labelsize=7)
plt.subplots_adjust(hspace=0.48)
```
